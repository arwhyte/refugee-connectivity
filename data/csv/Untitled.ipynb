```python
import pandas as pd

hourly_u = pd.read_csv('hourly_usage_per_camp.csv')
daily_u = pd.read_csv('daily_usage_per_camp.csv')
hourly_u.head()
```

```python
daily_u.head()
```

```python
daily_u = daily_u.drop(['Unnamed: 0'],axis=1)
```

```python
daily_total = hourly_u.groupby('date').sum()[['total_usage_kb']]
daily_total.head()
```

```python
final_daily = daily_u.join(daily_total,on='date')
final_daily.describe()
```

```python
final_daily[final_daily['total_usage_kb']==0]
```

```python
final_daily.to_csv('daily_usage_per_camp.csv')
```
